```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd
```

```python
text = df.skills[6]

# Create and generate a word cloud image:
wordcloud = WordCloud().generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
text = " ".join(review for review in df.skills)
print ("There are {} words in the combination of all review.".format(len(text)))
```

```python
stopwords = set(STOPWORDS)
stopwords.update(["developer", "development", "programmer",'processing',','])

# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)

# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud,interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
import collections
```

```python
filtered_words_spam = [skill for skill in text.split(',') if skill not in stopwords]
counted_words_spam = collections.Counter(filtered_words_spam)

word_count_spam = {}

for letter, count in counted_words_spam.most_common(35):
    word_count_spam[letter] = count
    
for i,j in word_count_spam.items():
        print('skill: {0}, count: {1}'.format(i,j))
```

```python
import matplotlib.pyplot as plt
top=sorted(word_count_spam.items(), key=lambda x:x[1],reverse=True)[:5] 
x,y=zip(*top)
plt.bar(x,y)
```

```python
wordcloud.to_file("python_developer.png")
```

```python
plt.savefig('python1.png')
```

```python
import seaborn as sns
from collections import Counter
counter=Counter(word_count_spam)
most=counter.most_common()

x, y= [], []
for word,count in most[:15]:
    
        x.append(word)
        y.append(count)
        
A=sns.barplot(x=y,y=x)
```

```python
A.figure.savefig('pythonsns.png')
```
