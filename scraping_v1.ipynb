```python
import requests
import bs4
from bs4 import BeautifulSoup
import pandas as pd
import time

URL = ("https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=python+developer&txtLocation=North+India ")
#conducting a request of the stated URL above:
page3 = requests.get(URL)
#specifying a desired format of “page” using the html parser - this allows python to read the various components of the page, rather than treating it as one long string.
soup = BeautifulSoup( page3.text, "html.parser")

```

```python
jobs=soup.find_all('li',class_='clearfix job-bx wht-shd-bx')
```

```python
for job in jobs:
    company=[]
    company1= job.find('h3',class_ ='joblist-comp-name')
    company.append(company1.text.strip())
    
```

```python
print(company)
```

```python
company=[]
for job in jobs:
    company1= job.find('h3',class_ ='joblist-comp-name')
    company.append(company1.text.strip())
    
    
```

```python
print(company)
```

```python
year_of_experience =[]
for job in jobs:
    years=job.find('ul', class_='top-jd-dtl clearfix')
    for years in jobs:
        year=years.find('li').text.replace('card_travel','')
        
        year_of_experience.append(year.strip())
        
        
        
    
```

```python
year_of_experience 
```

```python
skills=[]
for job in jobs:
    skills1=job.find('span',class_='srp-skills').text
    skills.append(skills1.strip())
```

```python
skills
```

```python
description=[]
for job in jobs:
    des=job.find('ul',class_='list-job-dtl clearfix')
    des1=des.li.text.replace('Job Description:','')
    
    
    description.append(des1.strip())
```

```python
description
```

```python
link=[]
for job in jobs:
    link1=job.header.h2.a['href']
    link.append(link1)
```

```python
link
```

```python
published_date=[]
for job in jobs:
    published_date1=job.find('span',class_='sim-posted').text
    published_date.append(published_date1.strip())
```

```python
published_date
```

```python
data_tuples = list(zip(company,year_of_experience,description,skills,link))
```

```python
data_tuples
```

```python
 df=pd.DataFrame(data_tuples, columns=['company','year_of_experience','description','skills','link'])
```

```python
df.to_csv(r' python_developer1.1.csv')
```
