```python
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

cv = CountVectorizer()
words = cv.fit_transform(data1['skills'])
sum_words = words.sum(axis=0)


words_freq = [(word, sum_words[0, idx]) for word, idx in cv.vocabulary_.items()]
words_freq = sorted(words_freq, key = lambda x: x[1], reverse = True)
frequency = pd.DataFrame(words_freq, columns=['word', 'freq'])

color = plt.cm.twilight(np.linspace(0, 1, 20))
frequency.head(20).plot(x='word', y='freq', kind='bar', figsize=(15, 7), color = color)
plt.title("HIGHLY REQUIRED SKILLS - Top 20")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer( max_features = 2000)
X = vectorizer.fit_transform(data1['skills']+data2['skills']+data3['skills'])

# getting the shape of X
print("Shape of X :", X.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
```

```python
true_k = 3
model = KMeans(n_clusters=8)
model.fit(X)
```

```python
print("Top terms per cluster:")

order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()

for i in range(true_k):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind])

```

```python
Y = vectorizer.transform(["python R mysql django"])
prediction = model.predict(Y)
print("Cluster number :", prediction)
```

```python
Y = vectorizer.transform([" R "])
prediction = model.predict(Y)
print("Cluster number :", prediction)
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd
```

```python
text1 = " ".join(review for review in data1.skills)
print ("There are {} words in the combination of all review.".format(len(text1)))
```

```python
text2 = " ".join(review for review in data2.skills)
print ("There are {} words in the combination of all review.".format(len(text2)))
```

```python
text3 = " ".join(review for review in data3.skills)
print ("There are {} words in the combination of all review.".format(len(text3)))
```

```python
text=text1+text2+text3
```

```python
text
```

```python
stopwords = set(STOPWORDS)
stopwords.update(["developer", "development", "programmer","processing",","])

# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)

# Display the generated image:
# the matplotlib way:
plt.figure(figsize = (15, 10))
plt.imshow(wordcloud,interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
import collections
filtered_words_spam = [skill for skill in text.split(',') if skill not in stopwords]
counted_words_spam = collections.Counter(filtered_words_spam)

word_count_spam = {}

for letter, count in counted_words_spam.most_common(35):
    word_count_spam[letter] = count
    
for i,j in word_count_spam.items():
        print('skill: {0}, count: {1}'.format(i,j))
```

```python
import matplotlib.pyplot as plt
top=sorted(word_count_spam.items(), key=lambda x:x[1],reverse=True)[:5] 
x,y=zip(*top)
plt.bar(x,y)
```

```python
import seaborn as sns
from collections import Counter
counter=Counter(word_count_spam)
most=counter.most_common()

x, y= [], []
for word,count in most[:15]:
    
        x.append(word)
        y.append(count)
        
sns.barplot(x=y,y=x)
```
